###Model training
For Model Training  data.yaml file is  must required ,and other parameter optional or you can do Hyperparameter tuning with other parameters like No. of epochs, Learning rate and many more

data.yaml file contains your
1. data folder path
2. train folder name
3. test/validation folder name
4. nc(no. of classes)
5. and all classes name

for ex:-

  1. path: /od_project/data/
  2. train: train
  3. val: test
  4. nc: 3
  5. names:
      1. 0: car
      2. 1: helmet
      3. 2: person

after that save this file(.yaml format) with any name
